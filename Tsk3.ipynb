# Question 6

What are the peak hours of sales?

```python
hourly_sales = (df.groupby("hour")["quantity"].sum().sort_index())  # Sort the index (hours) in ascending order

hourly_sales
```

```python
peak_hour = hourly_sales.idxmax()    # .idxmax() function is used to find the index (hour) of the maximum value in the hourly_sales Series. This gives us the hour during which the highest quantity of pizzas was sold.

print(f"Peak Hour: {peak_hour}:00")    # 00 indicates the start of the hour
```

```python
colors = plt.cm.Oranges(np.linspace(0.3, 1, len(hourly_sales)))    # Generate a list of colors from the Oranges colormap

plt.figure(figsize=(10,6))

plt.bar(hourly_sales.index, hourly_sales.values, color=colors)   # Create a bar plot with the specified colors

plt.title("Total Quantity Sold by Hour")   # Set the title of the plot
plt.xlabel("Hour of the Day")     # Set the x-axis label
plt.ylabel("Total Quantity Sold")   # Set the y-axis label

plt.xticks(hourly_sales.index)   # Set x-ticks to be the hours of the day

plt.show()    # Display the plot
```

Insight

The hourly sales analysis reveals that customer demand is highly concentrated during lunch hours, with the highest sales recorded at 12:00 PM (6,776 pizzas), followed closely by 1:00 PM (6,413 pizzas). A second surge in demand occurs between 5:00 PM and 7:00 PM, indicating a strong evening dining pattern. In contrast, sales are significantly lower during the early morning (9:00 AM–10:00 AM) and late evening (10:00 PM–11:00 PM).

These findings suggest that customer purchasing behavior is driven primarily by lunch and dinner meal times, making these periods the restaurant's most critical operating hours.

# Question 7

Find the total sales made on each day of the week. Which day of the week is when sales are made the most?

This question will help management to:

-Schedule staff appropriately.

-Plan inventory.

-Run promotions on slower days.

-Understand customer purchasing patterns.

```python
df.columns
```

```python
sales_by_day = df.groupby("day_name")["sales"].sum()    # group the DataFrame df by the "day_name" column and calculate the total sales for each day of the week. The result is a Series where the index is the day names and the values are the corresponding total sales.

sales_by_day
```

```python
days_order = [       # List of days of the week in the correct order
    "Monday",
    "Tuesday",
    "Wednesday",
    "Thursday",
    "Friday",
    "Saturday",
    "Sunday"
]
```

```python
sales_by_day = sales_by_day.reindex(days_order)   # reindex the sales_by_day Series to ensure that the days of the week are in the correct order (Monday to Sunday) based on the day_order list.
    
sales_by_day
```

```python
best_sales_day = sales_by_day.idxmax()   # .idxmax() function is used to find the index (day name) of the maximum value in the sales_by_day Series. This gives us the day of the week during which the highest total sales occurred.

print(f"Highest Sales Day: {best_sales_day}")
```

```python
plt.figure(figsize=(10,6))    # Create the figure

colors = plt.cm.Oranges(np.linspace(0.3, 1, len(sales_by_day)))    # Create shades of orange


plt.bar(sales_by_day.index, sales_by_day.values, color=colors)   # Plot the chart


plt.title("Total Sales by Day of the Week", fontsize=14, fontweight="bold")   # Add title and labels
plt.xlabel("Day of the Week")
plt.ylabel("Total Sales ($)")


plt.xticks(rotation=45)   # Rotate day labels


plt.tight_layout()    # Adjust layout


plt.show()      # Display chart
```

Insight

The weekly sales trend indicates that customer demand is highest towards the end of the workweek, particularly on Friday, likely due to customers dining out or ordering pizza to begin the weekend. The strong performance on Thursday, Friday, and Saturday presents an opportunity to maximize revenue through efficient staffing, inventory planning, and targeted promotions. Conversely, the lower sales on Sunday suggest an opportunity to introduce marketing campaigns or special offers to stimulate demand on slower days.

# Question 14


What Is the Average Order Value (AOV)

```python
# Calculate Average Order Value (AOV)
average_order_value = df["sales"].sum() / df["order_id"].nunique()

print(f"Average Order Value: ${average_order_value:.2f}")
```

```python
plt.figure(figsize=(4,6))

plt.bar(["Average Order Value"], [average_order_value], color="orange")

plt.title("Average Order Value (AOV)", fontsize=14, fontweight="bold")
plt.ylabel("Revenue ($)")

plt.show()
```
